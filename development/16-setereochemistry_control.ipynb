```python
from architector import convert_io_molecule, view_structures, build_complex
```

```python
view_structures("[O-][C@H](Cl)C[C@@H](Br)[O-]",labelinds=True)
```

```python
ligand = {'smiles':"[O-][C@H](Cl)C[C@@H](Br)[O-]",
          'coordList':[0,6],
          'ligType':'bi_cis'}
out = build_complex({'core':{'metal':'Fe','coreCN':6},
                     'ligands':[ligand]*1,
                     'parameters':{'enforce_ligand_internal_symmetry':True,
                                   'return_only_1':True}})
view_structures(out)
```
