```python
from architector import build_complex,view_structures,convert_io_molecule
out  = build_complex(
    {'core': {'metal': 'U', 'coreCN': 4}, 'ligands': 
    ['oxo','oxo','oxygen_edge'],
    'parameters': {'metal_ox': 6, 'metal_spin': 0, 
                   'assemble_method': 'UFF', 'full_method': 'UFF'}})
```

```python
view_structures(out)
```

```python
import copy
mol = convert_io_molecule(out[list(out.keys())[0]]['mol2string'])
mol1 = convert_io_molecule(out[list(out.keys())[0]]['mol2string'])
mol2 = copy.deepcopy(mol)
```

```python
view_structures(mol,labelinds=True)
```

```python
mol2.remove_atoms([1,2])
```

```python
mol.functionalize_3D(functional_groups=[mol2],
                     functionalization_inds=[[1, 2]],
                     bond_orders=[[1, 1]],
                     functional_group_mol_inds=[[0, 0]],
                     remove_hydrogens_when_adding=[[False,False]],
                     core_frozen=False,
                     xtb_opt=True)
```

```python
view_structures(mol,labelinds=True)
```

```python
view_structures(mol1,labelinds=True)
```

```python
mol.functionalize_3D(functional_groups=[mol1],
                     functionalization_inds=[[0, 0]],
                     bond_orders=[[1, 1]],
                     functional_group_mol_inds=[[1, 2]],
                     remove_hydrogens_when_adding=[[False,False]],
                     core_frozen=False,
                     xtb_opt=True)
```

```python
view_structures(mol,labelinds=True)
```

```python
mol.functionalize_3D(functional_groups=[mol1],
                     functionalization_inds=[[5, 5]],
                     bond_orders=[[1, 1]],
                     functional_group_mol_inds=[[1, 2]],
                     remove_hydrogens_when_adding=[[False,False]],
                     core_frozen=False,
                     xtb_opt=True)
```

```python
view_structures(mol,labelinds=True)
```

```python
view_structures(mol,w=400,h=400)
```

```python
from architector.io_arch_dock import add_non_covbound_species
```

```python
out = add_non_covbound_species(mol,
                         parameters={'species_list':[mol]*2,
                                     "species_add_copies":100, # Number "randomized" generations to do.
                                    "species_location_method":'random', # Use random for actiniyls 
                                    "species_relax":False, # Don't relax (none of the methods are super accurate)
                                    "species_intermediate_relax":False, # Don't relax intermediate
                                    "species_intermediate_method":'UFF', # Use UFF for speed
                                    "species_method":'UFF', # Use UFF for speed
                                    "debug":False}) # Turn on if you want to see what it's doing in the background })
```
